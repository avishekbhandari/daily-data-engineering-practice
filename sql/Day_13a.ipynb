###find all post which were reacted to with a heart

```python
%sql
SELECT p.post_id, p.poster, p.post_text, p.post_keywords, p.post_date, r.friend, r.reaction, r.date_day
FROM facebook_posts p
JOIN facebook_reactions r
ON p.post_id = r.post_id WHERE r.reaction = 'heart';
```
